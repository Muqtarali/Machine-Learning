```python
import os

def find_file(filename, search_path):
    for root, dirs, files in os.walk(search_path):
        if filename in files:
            return os.path.join(root, filename)
    return None

path = find_file('models.zip', '/content')
if path:
    print(f"🎉 Found it! The file is located at: {path}")
else:
    print("❌ The file 'models.zip' is not anywhere in your Colab environment.")
```
